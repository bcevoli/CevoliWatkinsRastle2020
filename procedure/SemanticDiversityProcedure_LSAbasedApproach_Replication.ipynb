```python
# Import used modules
import os
import sys
import time
import glob
import time
import numpy as np
import pandas as pd
import multiprocessing as mp
from datetime import datetime
from functools import partial
import Context_LSA_SemanticDiversityProcedure_Replication_workers
from Context_LSA_SemanticDiversityProcedure_Replication_utils import LogEntropyVectorizer 
```

```python
# Document-Term Matrix
```

```python
# Loop through each document of the BNC and get context information (for details see workers.getContexts)
# Set start time
start_time = time.time()
# Set parameters
lemmatization = True
# Get a list of all of the document directories
corpusPath =  #to add subfolder where the BNC is ("~BNC/Texts")
fileDirList = [path for path in glob.iglob(corpusPath+'**/**/*.xml', recursive=True)]
# Run it in parallel
if __name__ ==  '__main__': 
    num_processors = mp.cpu_count()
    print("Number of cores:", num_processors)
    func = partial(Context_LSA_SemanticDiversityProcedure_Replication_workers.getContexts, lemmatization)
    p = mp.Pool(processes = num_processors)
    corpusInfo = p.map(func, fileDirList)
    # Set end time
    end_time = time.time()
    print("Total time taken:", int((end_time - start_time)/60), "min")
```

```python
contextIDs = [context[1] for context in corpusInfo]
contextIDs = [contextID for context in contextIDs for contextID in context]
contextInfo = pd.DataFrame.from_dict(data = {'ContextID': contextIDs})
contextInfo = contextInfo[contextInfo["ContextID"].str.len()>1]
contextInfo.reset_index(inplace=True, drop=True)
contextInfo.reset_index(inplace=True, drop=False)
contextInfo.rename(columns={'index': 'contextIndex'}, inplace=True)
contextInfo["DocID"] = contextInfo["ContextID"].str[:-2]
contextInfo
```

```python
corpus = [contextText[0] for contextText in corpusInfo]
```

```python
# Number of documents 
len(corpus)
```

```python
# Spoken documents
spokenCorpus = [docs for docs in corpus if docs=="SPOKEN"]
print("Number of documents:", len(spokenCorpus))
print("Number of tokens:", sum([(len(wordList.split())) for docs in spokenCorpus for wordList in docs]))
```

```python
# Written documents
writtenCorpus = [docs for docs in corpus if docs!="SPOKEN"]
print("Number of documents:", len(writtenCorpus))
print("Number of tokens:", sum([(len(wordList.split())) for docs in writtenCorpus for wordList in docs]))
```

```python
# Number of subdivided documents (written docs only)
plainCorpus = [wordList for docs in writtenCorpus for wordList in docs]
len(plainCorpus)
```

```python
# Save document and term frequencies
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(min_df=40)
dtm_counts = cv.fit_transform(plainCorpus)
wordFreq = pd.DataFrame({"Word":cv.get_feature_names(), "WordCount": np.squeeze(np.asarray(dtm_counts.sum(axis=0)))})
wordFreq = wordFreq[wordFreq["WordCount"]>49]
wordFreq
```

```python
# Log Entropy Weighted Document-Term Matrix
vectorizer = LogEntropyVectorizer(norm=None, min_df=40, min_tf=50)
dtm = vectorizer.fit_transform(plainCorpus) 
dtm
```

```python
# Indices of context vectors in which words occur in
wordDocList = pd.DataFrame({"WordID": list(dtm.nonzero()[1]), "Contexts": list(dtm.nonzero()[0])}).groupby("WordID")["Docs"].unique().reset_index()
wordDocList["Word"] = vectorizer.get_feature_names()
wordDocList["nContexts"] = [len(docList) for docList in wordDocList["Contexts"]]
wordDocList
```

```python
def explode(df, lst_cols, fill_value='', preserve_index=False):
    # make sure `lst_cols` is list-alike
    if (lst_cols is not None
        and len(lst_cols) > 0
        and not isinstance(lst_cols, (list, tuple, np.ndarray, pd.Series))):
        lst_cols = [lst_cols]
    # all columns except `lst_cols`
    idx_cols = df.columns.difference(lst_cols)
    # calculate lengths of lists
    lens = df[lst_cols[0]].str.len()
    # preserve original index values    
    idx = np.repeat(df.index.values, lens)
    # create "exploded" DF
    res = (pd.DataFrame({
                col:np.repeat(df[col].values, lens)
                for col in idx_cols},
                index=idx)
             .assign(**{col:np.concatenate(df.loc[lens>0, col].values)
                            for col in lst_cols}))
    # append those rows that have empty lists
    if (lens == 0).any():
        # at least one list in cells is empty
        res = (res.append(df.loc[lens==0, idx_cols], sort=False)
                  .fillna(fill_value))
    # revert the original index order
    res = res.sort_index()
    # reset index if requested
    if not preserve_index:        
        res = res.reset_index(drop=True)
    return res
wordDocList = explode(wordDocList, ['Contexts'], fill_value='')
wordDocList.to_csv("data/wordContextList.csv", index=False)
wordDocList
```

```python
# Add number of contexts
wordFreq = pd.merge(wordFreq, wordDocList[["Word","nContexts"]], right_on="Word", left_on="Word", how="outer")
wordFreq
```

```python
# SVD
from sklearn.decomposition import TruncatedSVD

# Set start time
start_time = time.time()
# Singular Value Decomposition (SVD)
print("Shape of sparse matrix before SVD: ", dtm.shape)
svd = TruncatedSVD(n_components = 300)
dtm = svd.fit_transform(dtm)
print("Shape of sparse matrix after SVD: ", dtm.shape)
np.savetxt("data/LSAcontextVectors.npz", dtm, delimiter=",")
# Set end time
end_time = time.time()
print("Total time taken:", int((end_time - start_time)/60), "min")
```

```python
import matplotlib.pyplot as plt
plt.plot(np.log(svd.singular_values_))
plt.ylabel('Singular Values')
plt.show()
```

```python
# Loop through each word and compute semantic diversity (for details see workers.SemanticDiversity)
# Set start time
start_time = time.time()
# Singular Values Unweighting
svUnweighting = False
if svUnweighting==True:
    dtm = dtm/svd.singular_values_
# Run it in parallel
if __name__ ==  '__main__': 
    num_processors = mp.cpu_count()
    print("Number of cores:", num_processors)
    iterable = [(row["Word"], row["WordID"], row["Docs"]) for index, row in wordDocList.iterrows()]
    func = partial(Context_LSA_SemanticDiversityProcedure_Replication_workers.SaveDiversityInfo, dtm)
    p = mp.Pool(processes = num_processors)
    diversityInfo = p.map(func, iterable)
    # Set end time
    end_time = time.time()
    print("Total time taken: ", int((end_time - start_time)/60), "min")
replication = pd.DataFrame(list(diversityInfo), columns=["Word", "SemD", "cosMean"])
replication = pd.merge(replication, wordFreq, left_on="Word", right_on="Word", how="outer")
if (svUnweighting==True)&(lemmatization==True):
    replication.to_csv("data/diversityResults_LemmatizedCorpus_svUnweighted.csv.csv", index=False)
elif (svUnweighting==True)&(lemmatization==False):
    replication.to_csv("data/diversityResults_InflectedCorpus_svUnweighted.csv.csv", index=False)
elif (svUnweighting==False)&(lemmatization==False):
    replication.to_csv("data/diversityResults_LemmatizedCorpus_svUnweighted.csv.csv", index=False)
else:
    replication.to_csv("data/diversityResults_LemmatizedCorpus_svWeighted.csv", index=False)
```
